```python
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
#reading the dataset and splitting it
dp = pd.read_csv('train.csv')
X = dp.iloc[:,[2,4,5,6,7,9,11]].values
y = dp.iloc[:,1].values
```

```python
#filling in the missing values for age
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
X[:, 2:3] = imputer.fit_transform(X[:, 2:3])
```

```python
#encoding the gender column
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
X[:, 1] = le.fit_transform(X[:, 1])
```

```python
#encoding the embarked column
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers= [('encoder', OneHotEncoder(), [6])], remainder= 'passthrough')
X = np.array(ct.fit_transform(X))
```

```python
#training our model
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, random_state= 0)
```

```python
#scaling our model
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#building our ANN model
ann = tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=6, activation= 'relu'))
ann.add(tf.keras.layers.Dense(units=6, activation= 'relu'))
ann.add(tf.keras.layers.Dense(units=1, activation= 'sigmoid'))
```

```python
#compiling our ANN model
ann.compile(optimizer= 'adam', loss= 'binary_crossentropy', metrics= ['accuracy'] )
```

```python
#training our ANN model
ann.fit(X_train, y_train, batch_size= 16, epochs = 100)
```

```python
#predicting the test results
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5) #using > 0.5 returns a boolean result
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
test = pd.read_csv('test.csv')
A = test.iloc[:, [1,3,4,5,6,8,10]].values
```

```python
A[:, 2:3] = imputer.fit_transform(A[:, 2:3])
A[:, 1] = le.fit_transform(A[:, 1])
```

```python
A = np.array(ct.transform(A))
```

```python
prediction = (ann.predict(sc.transform(A)) > 0.5).astype(int)
```

```python
print(prediction)
```

```python
#creating the submission file
submission = pd.DataFrame({'PassengerId': test['PassengerId'], 'Survived': prediction.reshape(-1)})
submission.to_csv('Titanic_Submission.csv', index=False)
```
